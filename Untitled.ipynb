<h2> Companies with most layoffs </h2>

```python
sample = df[['company', 'total_laid_off']]
df2 = sample.groupby(['company']).sum() # adding together duplicate rows
Sorted = df2.sort_values( # highest to lowest
    by="total_laid_off",
    ascending=False
)
Sorted.head(11)
```

```python
#companies = Sorted['company']
new = Sorted[0:10]
new.hvplot.bar(y='total_laid_off', height=500)
```
